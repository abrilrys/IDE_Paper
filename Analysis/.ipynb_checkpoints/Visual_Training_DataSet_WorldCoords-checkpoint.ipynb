# Visual training data set 3D World Points

```python
%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import csv

gps_data = []
with open("../less_steps/controllers/my_controller/gps_hand.csv", "r", newline='') as gps_csvfile:
    gps_reader = csv.reader(gps_csvfile)
    for row in gps_reader:
        gps_data.append([float(value) for value in row[1:]])

x, y, z = zip(*gps_data)

# Convertir a listas
x = list(x)
y = list(y)
z = list(z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='blue', marker='o', s=1)  

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
plt.savefig('point_distribution.jpg')
```
