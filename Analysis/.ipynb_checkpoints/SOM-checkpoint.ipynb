# SOM Analysis

# Distribution of SOM Visual

```python
from minisom import MiniSom
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import random

sampleSize=5000

#normalize a data set
def min_max_normalize(x):
    min_val = np.min(x, axis=0)
    max_val = np.max(x, axis=0)
    normalized_x = (x - min_val) / (max_val - min_val)
    return normalized_x

# Load data
columns = ['Key','X', 'Y', 'Z']
data = pd.read_csv('..\less_steps\controllers\my_controller\gps_hand.csv', names=columns, sep=',', engine='python')


target = data['Key'].values

#print(target)
# Remove first column 
data = data[data.columns[1:]]
#print(data)
# Data normalization
data = min_max_normalize(data)
data = data.values

with open('..\less_steps\controllers\my_controller\somVisual.p', 'rb') as infile:
    somVisual = pickle.load(infile)

somVisual.topographic_error(data[:sampleSize])
```

```python
# plot som
w_x, w_y = zip(*[somVisual.winner(d) for d in data])
w_x = np.array(w_x)
w_y = np.array(w_y)

colors = ['#' + ''.join(random.choices('0123456789ABCDEF', k=6)) for _ in range(sampleSize)]

point_color = '#031cfc'

plt.figure(figsize=(19, 19))

plt.pcolor(somVisual.distance_map().T, cmap='bone_r', alpha=0.6)
plt.colorbar()

plt.xticks(np.arange(0, 70, 1),  rotation=90)
plt.yticks(np.arange(0, 70, 1))
plt.xlim(0, 70)
plt.ylim(0, 70)


plt.grid(color='black', linestyle='-', linewidth=0.5)  

for c in np.unique(target):
    idx_target = target == c
    plt.scatter(
        w_x[idx_target] + 0.5 + (np.random.rand(np.sum(idx_target)) - 0.5) * 0.8,
        w_y[idx_target] + 0.5 + (np.random.rand(np.sum(idx_target)) - 0.5) * 0.8,
        s=30,
        #c=colors[c - 1]
        c=point_color,
        alpha=0.8 
    )
plt.legend(loc='upper right')
plt.savefig('somVisual_plot.png', format='png', dpi=300)
plt.show()
```

```python
from minisom import MiniSom
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import random

sampleSize = 5000

# Normalize a dataset
def min_max_normalize(x):
    min_val = np.min(x, axis=0)
    max_val = np.max(x, axis=0)
    normalized_x = (x - min_val) / (max_val - min_val)
    return normalized_x

def plot_som(ax, file_path, som_path, title, point_color, columns):
    columns = columns
    data = pd.read_csv(file_path, names=columns, sep=',', engine='python')
    target = data['Key'].values
    data = data[data.columns[1:]]  # Remove first column
    data = min_max_normalize(data).values  # Data normalization

    with open(som_path, 'rb') as infile:
        som = pickle.load(infile)

    w_x, w_y = zip(*[som.winner(d) for d in data])
    w_x = np.array(w_x)
    w_y = np.array(w_y)

    ax.pcolor(som.distance_map().T, cmap='bone_r', alpha=0.6)
    ax.set_title(title)
    ax.set_xticks(np.arange(0, 70, 1), rotation=-90)
    ax.set_yticks(np.arange(0, 70, 1))
    ax.set_xlim(0, 70)
    ax.set_ylim(0, 70)
    ax.grid(color='black', linestyle='-', linewidth=0.5)

    for c in np.unique(target):
        idx_target = target == c
        ax.scatter(
            w_x[idx_target] + 0.5 + (np.random.rand(np.sum(idx_target)) - 0.5) * 0.8,
            w_y[idx_target] + 0.5 + (np.random.rand(np.sum(idx_target)) - 0.5) * 0.8,
            s=10,
            c=point_color,
            alpha=0.8
        )
    ax.tick_params(axis='x', labelsize=6) 
    ax.tick_params(axis='y', labelsize=6) 
        

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

plot_som(
    ax=axes[0],
    file_path='..\less_steps\controllers\my_controller\gps_hand.csv',
    som_path='..\less_steps\controllers\my_controller\somVisual.p',
    title='(a) Visual SOM',
    point_color = '#031cfc', 
    columns= ['Key', 'X', 'Y', 'Z']
)

# Plot (b)
plot_som(
    ax=axes[1],
    file_path='..\less_steps\controllers\my_controller\motor_angles.csv', 
    som_path='..\less_steps\controllers\my_controller\somAngles.p', 
    title='(b) Proprioceptive SOM',
    point_color = '#031cfc',
    columns = ['Key','RShoulderPitch', 'RShoulderRoll', 'RElbowYaw', 'RElbowRoll']
)

plt.tight_layout()
plt.savefig('side_by_side_som_visuals.png', format='png', dpi=300)
plt.show()
```
