# Implementation Less Steps

Get the 3D world coords for the learnt trajectories

```python
import json

# Load the JSON file
with open('../less_steps/controllers/my_controller/real_final_coords.json', 'r') as file:
    all_final_points = json.load(file)

for x in (all_final_points):
    print(f"Trajectory: {x}\n")
```

### 3D World Positions For Learnt Trayectories

```python
# %matplotlib notebook
from IPython.core.display import display, HTML
display(HTML("<style>div.output_scroll { height: 44em; }</style>"))

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

colors = ['red', 'green', 'blue', 'orange']
legends = ['goal 0', 'goal 1', 'goal 2', 'goal 3']

for i in range(len(all_final_points)):
    trajectory_to_plot = all_final_points[i]

    x = [coordinate[0] for coordinate in trajectory_to_plot]
    y = [coordinate[1] for coordinate in trajectory_to_plot]
    z = [coordinate[2] for coordinate in trajectory_to_plot]

    fig = plt.figure(figsize=(10, 7)) 
    ax = fig.add_subplot(111, projection='3d')

    for segment_index in range(len(colors)):
        start_idx = segment_index * 3
        end_idx = min(start_idx + 3, len(trajectory_to_plot))  

        if start_idx < len(trajectory_to_plot):
            x_segment = x[start_idx:end_idx + 1]
            y_segment = y[start_idx:end_idx + 1]
            z_segment = z[start_idx:end_idx + 1]

            ax.plot(x_segment, y_segment, z_segment, color=colors[segment_index], marker='o', label=legends[segment_index])

            for point_idx, (x_coord, y_coord, z_coord) in enumerate(zip(x_segment, y_segment, z_segment), start=start_idx):
                ax.text(x_coord, y_coord, z_coord + 0.0001, str(point_idx), color='blue')

    ax.plot(x, y, z, color='gray', linestyle='dotted', alpha=0.5)

    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_zlabel('Z Axis')
    
    ax.legend()
    
    plt.title(f"Trajectory {i}")
    plt.savefig(f"trajectory_{i}_wc_eng.png", dpi=300) 
    plt.show()  
```

### Associative error analisis

```python
import numpy as np
import matplotlib.pyplot as plt

def is_decreasing(lst):
    return all(lst[i] >= lst[i + 1] for i in range(len(lst) - 1))

distances_to_next_block = []
j=0
for coordinates in all_final_points:
    num_coords = len(coordinates)
    distances = []
    distancesBlocks=[]
    print(f"Trajectory {j}")
    j+=1
    for i in range(0, num_coords - 3, 3):  
        block = coordinates[i:i + 3]  
        next_coord = np.array(coordinates[i + 3])  
        
        for coord in block:
            dist = round(np.linalg.norm(np.array(coord) - next_coord), 4)
            distances.append(dist)
            distancesBlocks.append(dist)
            print(f"Coord: {coord}, compared to: {next_coord}, distance: {dist}")
        
        self_dist = round(np.linalg.norm(next_coord - next_coord), 4)
        distances.append(self_dist)
        distancesBlocks.append(self_dist)
        print(f"Reference Coord: {next_coord}, compared to itself, distance: {self_dist}")
        
        distancesBlocks = []
    distances_to_next_block.append(distances)  
    
   

plt.figure(figsize=(12, 8))
for i, distances in enumerate(distances_to_next_block):
    plt.plot(distances, label=f'Trajectory {i+1}')
plt.xlabel('Policy point index')
plt.ylabel('Distance to Next Goal Coordinate')
plt.title('Distance to Next Goal Coordinate for Each Trajectory')
plt.legend()
plt.grid()
plt.savefig(f"Error_To_Goals", dpi=300) 
plt.show()
```
